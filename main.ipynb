```python
from sentence_transformers import SentenceTransformer, util

import pymysql

conn = pymysql.connect(
       host="localhost",
       user="root",
       password="",
       database="toth",
       port=3306
)

cur = conn.cursor()

embedder = SentenceTransformer('all-MiniLM-L6-v2')

import json

def add_document(doc_data):
    # Step 1: Build context-rich text to embed
    context_text = (
        f"เนื้อหา: {doc_data['content']}\n"
        f"อาจารย์: {doc_data['teacher_name']} ({doc_data['teacher_subject']})\n"
        f"วันที่สอน: {doc_data['time_summit'].strftime('%Y-%m-%d')}\n"
        f"เวลาที่บันทึก: {doc_data['time_of_record'].strftime('%H:%M')}\n"
        f"ชั้นปี: ปี {doc_data['student_year']}, ห้อง {doc_data['student_room']}"
    )

    # Step 2: Create embedding from this full context
    embedding = embedder.encode(context_text).tolist()
    embedding_json = json.dumps(embedding)

    # Step 3: Save only original fields (no need to save the long string)
    cur.execute(
        """
        INSERT INTO `documents` 
        (`content`, `embedding`, `created_at`, `time_of_record`, `teacher_name`, `teacher_subject`, `student_year`, `student_room`)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
        """,
        (
            doc_data["content"],
            embedding_json,
            doc_data["time_summit"],
            doc_data["time_of_record"],
            doc_data["teacher_name"],
            doc_data["teacher_subject"],
            doc_data["student_year"],
            doc_data["student_room"]
        )
    )
    conn.commit()


    
from datetime import datetime, time

docs = [
    {
        "id": 1,
        "time_summit": datetime(2025, 6, 29).date(),
        "time_of_record": time(9, 30),
        "teacher_name": "อาจารย์สมชาย",
        "teacher_subject": "จิตวิทยา",
        "student_year": "2",
        "student_room": "201",
        "content": "ความรู้เรื่องสุขภาพจิต: การดูแลจิตใจของตัวเองในโลกยุคปัจจุบันมีความสำคัญไม่แพ้การดูแลสุขภาพกาย"
    },
    {
        "id": 2,
        "time_summit": datetime(2025, 6, 29).date(),
        "time_of_record": time(10, 15),
        "teacher_name": "อาจารย์วิทยา",
        "teacher_subject": "เทคโนโลยี",
        "student_year": "3",
        "student_room": "302",
        "content": "ความรู้ด้านเทคโนโลยี: การเข้าใจเทคโนโลยีพื้นฐาน เช่น ปัญญาประดิษฐ์ (AI) หรือการใช้แอปพลิเคชันต่าง ๆ ช่วยเพิ่มโอกาสในการทำงาน"
    },
    {
        "id": 3,
        "time_summit": datetime(2025, 6, 29).date(),
        "time_of_record": time(11, 0),
        "teacher_name": "อาจารย์สุภาพ",
        "teacher_subject": "การเงิน",
        "student_year": "4",
        "student_room": "401",
        "content": "ความรู้ทางการเงิน: การบริหารจัดการเงินอย่างมีประสิทธิภาพช่วยให้มีความมั่นคงทางการเงินในระยะยาว"
    },
    {
        "id": 4,
        "time_summit": datetime(2025, 6, 29).date(),
        "time_of_record": time(13, 30),
        "teacher_name": "อาจารย์นิรันดร์",
        "teacher_subject": "ภาษาศาสตร์",
        "student_year": "1",
        "student_room": "101",
        "content": "ความรู้ทางภาษา: การเรียนรู้ภาษาใหม่ช่วยเปิดโลกทัศน์ และสร้างโอกาสใหม่ ๆ ในการทำงานและการเดินทาง"
    },
    {
        "id": 5,
        "time_summit": datetime(2025, 6, 29).date(),
        "time_of_record": time(14, 45),
        "teacher_name": "อาจารย์ธีรชัย",
        "teacher_subject": "สิ่งแวดล้อม",
        "student_year": "2",
        "student_room": "205",
        "content": "ความรู้เรื่องสิ่งแวดล้อม: การเข้าใจปัญหาสิ่งแวดล้อม เช่น ภาวะโลกร้อน และการใช้พลังงานอย่างยั่งยืนเป็นหน้าที่ของพลเมืองโลก"
    }
]

newdata = [
    {
           "time_summit": datetime(2025, 6, 29).date(),
    "time_of_record": time(0, 50),
    "teacher_name": "อาจารย์วรเพียร์",
    "teacher_subject": "ชีวะวิทยา",
    "student_year": "3",
    "student_room": "301",
    "content": "สวัสดีนักเรียนทุกคนครับ/ค่ะ วันนี้ครูมีคำถามเปิดคาบเรียนให้นักเรียนลองคิดกันดูนะครับ/ค่ะ…ทุกคนคิดว่า ‘ร่างกายของเราสร้างขึ้นมาจากอะไร?’ ถูกต้องครับ/ค่ะ คำตอบก็คือ ‘เซลล์’ นั่นเอง วันนี้เราจะมาเรียนรู้เรื่องเซลล์ของสิ่งมีชีวิต ซึ่งเป็นหน่วยพื้นฐานที่เล็กที่สุดของชีวิต เซลล์นั้นมีหลายชนิด แต่หลักๆ เราสามารถแบ่งได้เป็น 2 แบบใหญ่ๆ คือ เซลล์โปรคาริโอต ซึ่งเป็นเซลล์ที่ไม่มีเยื่อหุ้มนิวเคลียส เช่น แบคทีเรีย กับอีกแบบคือ เซลล์ยูคาริโอต ซึ่งมีนิวเคลียสชัดเจน เช่น เซลล์พืชและเซลล์สัตว์ ในเซลล์จะมีออร์แกเนลล์มากมาย เช่น ไมโทคอนเดรีย ที่เป็นแหล่งสร้างพลังงาน ไรโบโซม ที่ใช้สร้างโปรตีน เอนโดพลาสมิกเรติคูลัม ที่เกี่ยวข้องกับการขนส่งสาร และอีกมากมาย ซึ่งแต่ละส่วนจะทำงานร่วมกันเพื่อให้เซลล์มีชีวิตอยู่ได้และทำหน้าที่ต่างๆ ได้อย่างเหมาะสม สุดท้ายนี้ อย่าลืมนะครับ/ค่ะว่าเซลล์คือรากฐานสำคัญที่สุดของสิ่งมีชีวิตทุกชนิด จบคาบเรียนวันนี้เพียงเท่านี้ ขอบคุณนักเรียนทุกคนครับ/ค่ะ"
    },
    {
           "time_summit": datetime(2025, 6, 27).date(),
    "time_of_record": time(1, 40),
    "teacher_name": "อาจารย์วรเพียร์",
    "teacher_subject": "ชีวะวิทยา",
    "student_year": "3",
    "student_room": "301",
    "content": "สวัสดีนักเรียนครับ/ค่ะ วันนี้ครูอยากชวนนักเรียนทุกคนคิดดูว่า เราเกิดมาได้อย่างไร? หรือสิ่งมีชีวิตอื่นๆ อย่างแมว หมา หรือแมลง มาจากไหน? คำตอบก็คือ การสืบพันธุ์นั่นเองครับ/ค่ะ วันนี้เราจะมาเรียนรู้เรื่องการสืบพันธุ์ของสิ่งมีชีวิต ซึ่งแบ่งได้เป็น 2 แบบหลักๆ คือ การสืบพันธุ์แบบไม่อาศัยเพศ เช่น การแตกหน่อ การแบ่งตัว หรือการงอกใหม่ ซึ่งสิ่งมีชีวิตจะสามารถเพิ่มจำนวนตัวเองได้โดยไม่ต้องมีเพศตรงข้าม กับแบบที่สองคือ การสืบพันธุ์แบบอาศัยเพศ ซึ่งต้องอาศัยเซลล์สืบพันธุ์ของเพศผู้และเพศเมียมาผสมกัน เช่น อสุจิกับไข่ เมื่อตัวอสุจิเข้าผสมกับไข่จะเกิดการปฏิสนธิและเจริญเติบโตเป็นตัวอ่อน ซึ่งในมนุษย์ก็เป็นแบบนี้เช่นกัน การสืบพันธุ์แบบอาศัยเพศจะช่วยให้เกิดความหลากหลายทางพันธุกรรม ขณะที่แบบไม่อาศัยเพศทำได้รวดเร็วกว่าและไม่ต้องใช้คู่ สรุปง่ายๆ คือ การสืบพันธุ์คือกระบวนการสำคัญที่ช่วยให้สิ่งมีชีวิตดำรงเผ่าพันธุ์ต่อไปได้ จบคาบวันนี้ ขอบคุณครับ/ค่ะ"
    },
    {
           "time_summit": datetime(2025, 6, 26).date(),
    "time_of_record": time(1, 40),
    "teacher_name": "อาจารย์วรเพียร์",
    "teacher_subject": "ชีวะวิทยา",
    "student_year": "3",
    "student_room": "301",
    "content": "สวัสดีนักเรียนครับ/ค่ะ เคยรู้สึกไหมว่าเวลาวิ่งแล้วเหนื่อย หรือเวลาทานอาหารแล้วรู้สึกอิ่ม นั่นเป็นเพราะมีหลายระบบในร่างกายที่ทำงานร่วมกัน วันนี้เราจะมาเรียนเรื่องระบบร่างกายของมนุษย์ ซึ่งเป็นหัวใจสำคัญของการดำรงชีวิต ระบบแรกคือระบบไหลเวียนโลหิต ที่มีหัวใจทำหน้าที่สูบฉีดเลือดไปทั่วร่างกาย ระบบที่สองคือระบบหายใจ ซึ่งช่วยนำออกซิเจนเข้าสู่ร่างกายและกำจัดคาร์บอนไดออกไซด์ออกไป ระบบที่สามคือระบบย่อยอาหาร ทำหน้าที่ย่อยอาหารให้เป็นสารอาหารที่ร่างกายสามารถนำไปใช้ได้ ระบบประสาทควบคุมการทำงานของร่างกายและการตอบสนองต่อสิ่งเร้า และยังมีระบบขับถ่าย ระบบสืบพันธุ์ และอีกมากมาย ทุกระบบล้วนเชื่อมโยงกัน ทำงานประสานกันอย่างเป็นระบบเพื่อให้เรามีชีวิตอยู่ได้อย่างปกติ ดังนั้น ถ้าระบบใดระบบหนึ่งมีปัญหา ร่างกายก็จะทำงานผิดปกติไปด้วย สรุปคือ ระบบต่างๆ ในร่างกายมนุษย์ล้วนมีหน้าที่เฉพาะของตัวเอง และเมื่อทำงานร่วมกันอย่างสมดุล ก็ทำให้เรามีสุขภาพที่ดี จบคาบเรียนวันนี้ครับ/ค่ะ ขอบคุณทุกคนครับ/ค่ะ"
    },
    {
           "time_summit": datetime(2025, 6, 23).date(),
    "time_of_record": time(0, 50),
    "teacher_name": "อาจารย์วรเพียร์",
    "teacher_subject": "ชีวะวิทยา",
    "student_year": "3",
    "student_room": "301",
    "content": "สวัสดีนักเรียนครับ/ค่ะ เคยสังเกตไหมว่าทำไมเราถึงหน้าตาคล้ายพ่อแม่ หรือพี่น้องถึงมีทั้งความเหมือนและความต่าง? คำตอบอยู่ที่พันธุกรรมครับ/ค่ะ วันนี้เราจะเรียนเรื่องพันธุกรรมและดีเอ็นเอ ซึ่งเป็นรหัสชีวิตที่ถ่ายทอดลักษณะจากรุ่นหนึ่งไปสู่อีกรุ่นหนึ่ง พันธุกรรมถูกควบคุมโดยดีเอ็นเอ ซึ่งเป็นสารเคมีรูปทรงเกลียวคู่ที่อยู่ภายในนิวเคลียสของเซลล์ ทุกเซลล์ในร่างกายจะมีดีเอ็นเอที่บอกว่าเราจะสูงเท่าไร มีสีตาแบบไหน หรือมีกรุ๊ปเลือดอะไร ดีเอ็นเอจะถูกรวมตัวเป็นหน่วยเล็กๆ ที่เรียกว่า ยีน และยีนจะจัดอยู่บนโครโมโซม ซึ่งมนุษย์มีทั้งหมด 46 แท่ง หรือ 23 คู่ นักวิทยาศาสตร์ชื่อเมนเดลเป็นคนแรกที่ค้นพบกฎการถ่ายทอดลักษณะทางพันธุกรรมจากการทดลองกับต้นถั่ว ทำให้เราสามารถใช้ความรู้เหล่านี้ในการพยากรณ์ลักษณะของลูกในอนาคตได้ เช่น การใช้แผนผัง Punnett Square เพื่อทำนายลักษณะเด่นและด้อย สรุปคือ พันธุกรรมและดีเอ็นเอคือพื้นฐานสำคัญที่บอกว่าเราจะมีลักษณะอย่างไร ซึ่งสิ่งนี้สืบทอดมาจากพ่อแม่สู่ลูก จบคาบเรียนวันนี้ครับ/ค่ะ ขอบคุณครับ/ค่ะ"
    },
    {
           "time_summit": datetime(2025, 6, 21).date(),
    "time_of_record": time(0, 50),
    "teacher_name": "อาจารย์วรเพียร์",
    "teacher_subject": "ชีวะวิทยา",
    "student_year": "3",
    "student_room": "301",
    "content": "สวัสดีนักเรียนครับ/ค่ะ เคยสงสัยไหมว่าในโลกนี้มีสิ่งมีชีวิตกี่ชนิด? พืช สัตว์ เห็ดรา แบคทีเรีย ล้วนเป็นสิ่งมีชีวิตที่มีความแตกต่างกันอย่างมาก วันนี้เราจะมาเรียนเรื่องความหลากหลายของสิ่งมีชีวิตและการจำแนก สิ่งมีชีวิตทั้งหมดบนโลกถูกจำแนกออกเป็นกลุ่มใหญ่ๆ ที่เรียกว่า อาณาจักร เช่น อาณาจักรพืช อาณาจักรสัตว์ อาณาจักรฟังไจ อาณาจักรโปรติสต์ และอาณาจักรแบคทีเรีย การจำแนกนี้ช่วยให้เราสามารถศึกษาและเข้าใจสิ่งมีชีวิตต่างๆ ได้ง่ายขึ้น แต่ละกลุ่มมีลักษณะเฉพาะ เช่น พืชสามารถสังเคราะห์แสงได้ ส่วนสัตว์ไม่สามารถทำได้ ฟังไจไม่มีคลอโรฟิลล์และดูดซึมอาหารจากสิ่งอื่นๆ นอกจากนี้ เรายังมีระบบการตั้งชื่อสิ่งมีชีวิตแบบทวินาม โดยใช้ชื่อสกุลและชื่อชนิด เช่น มนุษย์มีชื่อทางวิทยาศาสตร์ว่า Homo sapiens การมีความหลากหลายของสิ่งมีชีวิตเป็นสิ่งสำคัญต่อความสมดุลของธรรมชาติ เพราะแต่ละสิ่งมีชีวิตล้วนมีบทบาทต่อห่วงโซ่อาหารและระบบนิเวศ สรุปคือ ความหลากหลายของสิ่งมีชีวิตทำให้ธรรมชาติมีความสมดุล และการจำแนกสิ่งมีชีวิตช่วยให้เราศึกษาได้อย่างเป็นระบบ จบคาบเรียนวันนี้ครับ/ค่ะ ขอบคุณครับ/ค่ะ"
    }
]

for doc_data in newdata:
       add_document(doc_data)
       
cur.close()
conn.close()
```

```python
import pymysql
import json
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


def qeury_database(query, k=2):
    # Step 1: Encode query
    query_embedding = np.array(embedder.encode(query)).reshape(1, -1)

    # Step 2: Connect to MySQL
    conn = pymysql.connect(
        host="localhost",
        user="root",
        password="",
        database="toth",
        port=3306
    )
    cur = conn.cursor()

    # Step 3: Fetch all documents and their embeddings
    cur.execute("SELECT content, embedding, created_at, time_of_record, teacher_name, teacher_subject, student_year, student_room FROM documents")
    rows = cur.fetchall()

    results = []

    for row in rows:
        content, embedding_str, created_at, time_of_record, teacher_name, teacher_subject, student_year, student_room = row

        try:
            doc_embedding = np.array(json.loads(embedding_str)).reshape(1, -1)
            similarity = cosine_similarity(
                query_embedding, doc_embedding)[0][0]
            results.append((
                similarity,
                content,
                created_at,
                time_of_record,
                teacher_name,
                teacher_subject,
                student_year,
                student_room
            ))
        except Exception as e:
            print(f"Error parsing embedding: {e}")

    # Step 4: Sort results by similarity score (descending)
    results.sort(key=lambda x: x[0], reverse=True)

    # Step 5: Return top k results
    return results[:k]


# Run the query
for res in qeury_database("เซลล์"):
    print(res)
```
